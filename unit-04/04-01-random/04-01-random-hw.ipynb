```python
### BEGIN SOLUTION
n_max=0 #this is a global variable

def nextRandomInteger(seed,multiplier,increment,modulus):
    global n_max #our function needs to know what the last value of n is
    n_max = n_max + 1 #increment to the next value of n

    n=0 #start at n=0
    r=seed
    while n<n_max:
        r=(multiplier*r + increment) % modulus
        n = n + 1 #increment n
    
    #return r for n_max
    return r

#linear congruential method
r0 = 8697 #seed
mult = 229
inc = 0
mod = 349

r1 = nextRandomInteger(r0,mult,inc,mod) #the list of variables when we call the function are called arguments.
print("r_1 = ", r1)
r2 = nextRandomInteger(r0,mult,inc,mod) #the list of variables when we call the function are called arguments.
print("r_2 = ", r2)
r3 = nextRandomInteger(r0,mult,inc,mod) #the list of variables when we call the function are called arguments.
print("r_3 = ", r3)

#floating-point numbers
f1 = r1/mod
print("float1 = ", f1)
f2 = r2/mod
print("float2 = ", f2)
f3 = r3/mod
print("float3 = ", f3)

print()
print("The maximum number of random integers is ", mod)
### END SOLUTION
```

## Exercise 4

1. Plot the function

  $$f(x)=\sqrt{\cos^2(x)+1}$$

1. Write a loop that creates a random number for x between 0 and 2.0 and a random number for y between 0 and 1.5. Check if $y$ is less than $f(x)$ (under the curve) or greater than $f(x)$ (over the curve). If it's under the curve, increment a variable to count the data point. Run the loop 1000 times. Compute the ratio of the number of points under the curve to the total number of points. Use this method to estimate the area under the curve.

1. Store the random data points in a list and plot the points on the same graph as the function. Your picture should look similar to Figure 9.2.2, but with a different function plotted.

```python
### BEGIN SOLUTION

# create data points for the function
x_data = np.linspace(0,2,100)
y_data = np.sqrt((np.cos(x_data))**2+1)

# plot the function
fit = plt.figure()
plt.title('Dartboard')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x_data,y_data,'b-')
plt.ylim(0,1.5)
plt.show()


# dart simulation
N = 1000 #number of times to throw a dart

x_darts = [] #list to store x values for darts
y_darts = [] #list to store y values for darts

hits = 0 #a counter to count the number of times a dart is below the function

for i in range(N):
    x = 2*rand.random()
    y = 1.5*rand.random()
    
    x_darts.append(x)
    y_darts.append(y)
    
    y_func = np.sqrt((np.cos(x))**2+1)
    
    if y<y_func:
        hits = hits + 1

ratio = hits/N
ratio_percent = hits/N*100

A_rect = 1.5*2
A = ratio*A_rect
        
print("The number of darts below the function is ", hits)

print("The ratio of darts below the function is ", ratio)

print("The percentage of darts below the function is ", ratio_percent)

print("The area of the rectangle is ", A_rect)

print("The estimated area below the curve is ", A)

plt.title('Dartboard')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x_data,y_data,'b-')
plt.plot(x_darts,y_darts,'r.')
plt.ylim(0,1.5)
plt.show()
### END SOLUTION
```

```python
### BEGIN SOLUTION

#domain 
xi = 2
xf = 3
yi = xi**2
yf = xf**2

# create data points for the function
x_data = np.linspace(xi,xf,100)
y_data = x_data**2 #the function y(x)

# plot the function
fit = plt.figure()
plt.title('Dartboard')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x_data,y_data,'b-')
plt.show()


# dart simulation
N = 1000 #number of times to throw a dart

x_darts = [] #list to store x values for darts
y_darts = [] #list to store y values for darts

hits = 0 #a counter to count the number of times a dart is below the function

for i in range(N):
    x = rand.uniform(xi, xf)
    y = rand.uniform(yi, yf)
    
    x_darts.append(x)
    y_darts.append(y)
    
    y_func = x**2
    
    if y<y_func:
        hits = hits + 1

ratio = hits/N
ratio_percent = hits/N*100

A_rect = (xf-xi)*(yf-yi) #area of rectangle bounded by xmin, xmax, ymin, ymax
A = ratio*A_rect
        
print("The number of darts below the function is ", hits)

print("The ratio of darts below the function is ", ratio)

print("The percentage of darts below the function is ", ratio_percent)

print("The area of the rectangle is ", A_rect)

print("The estimated area below the curve is ", A)

plt.title('Dartboard')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x_data,y_data,'b-')
plt.plot(x_darts,y_darts,'r.')
plt.show()
### END SOLUTION
```
